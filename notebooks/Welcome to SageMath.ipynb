## Sage as a Calculator

The Sage command line has a sage: prompt; you do not have to add it. If you use the Sage notebook, then put everything after the sage: prompt in an input cell, and press shift-enter to compute the corresponding output.

```python
3+5
```

The caret symbol means “raise to a power”.

```python
57.1 ^ 100
```

We compute the inverse of a $2×2$ matrix in Sage.

```python
matrix([[1,2], [3,4]])^(-1)
```

Here we integrate a simple function.

```python
x = var('x')   # create a symbolic variable
integrate(sqrt(x)*sqrt(1+x), x)
1/4*((x + 1)^(3/2)/x^(3/2) + sqrt(x + 1)/sqrt(x))/((x + 1)^2/x^2 - 2*(x + 1)/x + 1) - 1/8*log(sqrt(x + 1)/sqrt(x) + 1) + 1/8*log(sqrt(x + 1)/sqrt(x) - 1)
```

This asks Sage to solve a quadratic equation. The symbol == represents equality in Sage.

```python
a = var('a')
S = solve(x^2 + x == a, x); S
[x == -1/2*sqrt(4*a + 1) - 1/2, x == 1/2*sqrt(4*a + 1) - 1/2]
```

The result is a list of equalities.

```python
S[0].rhs()
```

## Power Computing with Sage

First we create a $500×500$ matrix of random numbers.

```python
m = random_matrix(RDF,500)
```

It takes Sage a few seconds to compute the eigenvalues of the matrix and plot them.

```python
e = m.eigenvalues()  #about 2 seconds
w = [(i, abs(e[i])) for i in range(len(e))]
show(points(w))
```

Thanks to the GNU Multiprecision Library (GMP), Sage can handle very large numbers, even numbers with millions or billions of digits.

```python
factorial(100)
```

```python
n = factorial(1000000)  #about 2.5 seconds
```

This computes at least 100 digits of $\pi$.

```python
N(pi, digits=100)
```

This asks Sage to factor a polynomial in two variables.

```python
R.<x,y> = QQ[]
F = factor(x^99 + y^99)
F
```

```python
F.expand()
```

Sage takes just under 5 seconds to compute the numbers of ways to partition one hundred million as a sum of positive integers.

```python
z = Partitions(10^8).cardinality() #about 4.5 seconds
str(z)[:40]
```
